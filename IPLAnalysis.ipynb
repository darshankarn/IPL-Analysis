```python
# isin function
# Q.Find most destructive death over batsman in the history of ipl
# strike rate = (number of runs/number of balls played)/100
# death over is 16-20 over
# min balls played in death over is 200
```

```python
death_over = deliveries[deliveries['over']>15]
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
mask = all_batsman>200
death_over_batsman = all_batsman[mask].index.tolist()
death_over_batsman
```

```python
final_batsman_list = death_over[death_over['batsman'].isin(death_over_batsman)]
final_batsman_list
```

```python
#runs scored and balls played
runs = final_batsman_list.groupby('batsman')['batsman_runs'].sum()
balls = final_batsman_list.groupby('batsman')['batsman_runs'].count()

sr = (runs/balls)*100
sr.sort_values(ascending = False).head(10)
```

```python
new = data.merge(deliveries,left_on='id',right_on='match_id')
final_result = new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates('season',keep='first').sort_values('season')
final_result.reset_index()[['season','batsman']]
```
